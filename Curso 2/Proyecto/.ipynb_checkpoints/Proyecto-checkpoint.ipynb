## Entendimiento de los datos
El siguiente conjunto de datos fue extraído del repositorio https://www.kaggle.com/datasets/anshtanwar/jellyfish-types/data, el cual cuenta con 900 imágenes sobre distintas especies de medusas. El objetivo del proyecto es aplicar las técnicas de Machine Learning durante el curso y poder crear un clasificador de medusas a partir de las imágenes brindadas.

Son 900 imágenes que se encuentran divididas en 6 categorías, cada una con 150 imágenes organizadas en carpetas con el respectivo nombre de la clase a la cual pertenecen cada medusa.

Las clases a la las que pertenecen las medusas del conjunto de datos son las siguientes:

1. Barrel jellyfish (Rhizostoma pulmo): La medusa más grande que se encuentra en aguas británicas, con una campana que puede crecer hasta 90 cm de diámetro. Se alimenta de plancton y peces pequeños atrapándolos con sus tentáculos.
2. Blue jellyfish (Cyanea lamarckii): Medusas de gran tamaño que pueden crecer hasta 30 cm de diámetro. Se alimenta de plancton y peces pequeños atrapándolos con sus tentáculos.
3. Compass jellyfish (Chrysaora hysoscella): El nombre se debe a las marcas marrones en su campana que se asemejan a una rosa de los vientos. Se alimenta de plancton y peces pequeños atrapándolos con sus tentáculos.
4. Lion’s mane jellyfish (Cyanea capillata): La medusa más grande del mundo, con una campana que puede crecer hasta 2 metros de diámetro y tentáculos que pueden alcanzar hasta 30 metros de largo. Se alimenta de plancton y peces pequeños atrapándolos con sus tentáculos.
5. Mauve stinger (Pelagia noctiluca): Pequeña medusa con largos tentáculos y estructuras verrugosas en su campana repletas de células urticantes. Se alimenta de otras pequeñas medusas y ascidias oceánicas.
6. Moon jellyfish (Aurelia aurita): Medusa común con cuatro gónadas en forma de herradura visibles a través de la parte superior de su campana translúcida. Se alimenta recolectando medusas, plancton y moluscos con sus tentáculos.


```python
import cv2
import numpy as np
import os
from tqdm import tqdm
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import vlfeat
```

```python
class_name = ['barrel_jellyfish', 'blue_jellyfish', 'compass_jellyfish', 'lions_mane_jellyfish', 'mauve_stinger_jellyfish', 'Moon_jellyfish']
class_names_label = {class_name: i for i, class_name in enumerate(class_name)}
count_class = len(class_name)
IMAGE_SIZE = (240, 240)
SEED = 17
```

```python
def load_data(class_name):

    RGB = []
    gray = []
    labels = []
    #Carpetas que contienen las fotos
    for folder in class_name:

        print(f'Cargando "{folder}"')
        #Captura la clase y la transforma a volor numerico cuando se leen las fotos de una carpeta
        label = class_names_label[folder]

        #Lee las carpetas en una foto
        for file in tqdm(os.listdir(folder)):

            #Une el nombre de la carpeta con el nombre del archivo
            img_path = os.path.join(folder, file)
            
            #Lee la foto
            image = cv2.imread(img_path)
            
            #Cambia el formato de la foto de BGR a RGB
            image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            
            #Cambia el formato de la foto de BGR a gris
            image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
                
            #Agrega la foto cargada y procesada a la lista
            RGB.append(image_RGB)
            gray.append(image_gray)
            #Agrega el numero de la clase a la cual pertenece
            labels.append(label)
    #Agrupa todas las clases en un grupo
    labels = np.array(labels, dtype = 'int32')

    return RGB, gray, labels
```

```python
def display_image(index,data,label):
    plt.figure()
    plt.imshow(data[index])
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.title(f"#{index}: {class_name[label[index]]}")
    plt.show()
```

```python
def list_confusion_matrix(cm,classes):
  df = pd.DataFrame(data = cm,
                    index = pd.MultiIndex.from_product([['Valor real'], classes]),
                    columns = pd.MultiIndex.from_product([['Valor predicho'], classes]))

  return df
```

Las imágenes de las medusas son de distintos tamaños. El atributo de la clase a la cual pertenece es de una vez capturado cuando se listan los archivos para cada carpeta

```python
RGB, gray, y = load_data(class_name)
```

Una fotografía al ser leída y transformada a formato RGB se vuelve una matriz de 3 dimensiones para cada píxel, en donde subíndice indica la intensidad del canal RGB que componen el píxel.

```python
display_image(0, RGB, y)
```

## Preparación de los datos

Para poder usar las imágenes en un modelo de aprendizaje es necesario realizar una transformación, para ello, vamos a usar el algoritmo SIFT (Scale-invariant feature transform) que detecta, describe y hace coincidir las características locales en una imagen. Primero se debe extraer los puntos claves SIFT y almacenar los en una base de datos. Un objeto es reconocido como una nueva imagen comparando individualmente cada característica de la nueva imagen con la base de datos y encontrando características candidatas coincidentes basada en la distancia euclidiana de su vector de características.

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

class SIFT_transformer(BaseEstimator, TransformerMixin):
    
    def __init__(self, random_state=None, n_clusters=None):
        self.random_state = random_state
        self.n_clusters = n_clusters
        print('SIFT_transformer')

    def extract_SIFT_features(self, images):
        image_descriptors = []
        sift = cv2.SIFT_create()
        for image in images:
            _, des = sift.detectAndCompute(image, None)
            image_descriptors.append(des)
        return image_descriptors

    def kmean_BoW(self, all_descriptors, n_clusters):
        
        kmeans = KMeans(n_clusters=n_clusters,
                        random_state= self.random_state
                       )
        kmeans.fit(all_descriptors)
      
        return kmeans.cluster_centers_

    def create_feature_BoW(self, descriptors, BoW, num_cluster):
        
        X_features = []
        for i in range(len(descriptors)):
            features = np.array([0]*num_cluster)
            
            if descriptors[i] is not None:
                distance = cdist(descriptors[i], BoW)
                argmin = np.argmin(distance, axis = 1)
                for j in argmin:
                    features[j] += 1
            X_features.append(features)
        return X_features

    def fit(self, X, y=None):
        
        return self
    
    def transform(self, X, y=None, n_clusters=8):

        X_ = X.copy()
        desctiptors = self.extract_SIFT_features(X_)
        
        all_descriptors = []

        for descriptor in desctiptors:
            if descriptor is not None:
                for des in descriptor:
                    all_descriptors.append(des)

        BoW = self.kmean_BoW(all_descriptors, n_clusters)

        X_features = self.create_feature_BoW(desctiptors, BoW, n_clusters)

        return X_features

sift_RGB = SIFT_transformer(random_state=SEED, n_clusters=100)
RGB = sift_RGB.transform(RGB)

sift_gray = SIFT_transformer(random_state=SEED, n_clusters=100)
gray = sift_gray.transform(gray)
```

### División del conjunto de datos en entrenamiento, validación y prueba
Para la creación de los subconjuntos de se usará una proporción de 70% para el conjunto de entrenamiento y un 30% para el conjunto de pruebas


```python
from sklearn.model_selection import train_test_split


RGB_train, RGB_test, RGB_y_train, RGB_y_test = train_test_split(RGB,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=SEED,
                                                    stratify=y
                                                   )

gray_train, gray_test, gray_y_train, gray_y_test = train_test_split(gray,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=SEED,
                                                    stratify=y
                                                   )

_, train_counts = np.unique(RGB_y_train, return_counts=True)
_, test_counts = np.unique(RGB_y_test, return_counts=True)
pd.DataFrame({'train': train_counts,
              'test': test_counts
             }).plot.bar()
```

## Modelamiento
En la preparación de los datos se crearon 2 conjuntos de datos en donde uno corresponde a las imágenes en formato RGB y el otro a las imágenes en escala de grises, con el fin de saber formato presenta un mejor desempeño a la hora de clasificar las medusas. También se usarán 2 modelos de aprendizaje máquina de vectores de soporte y Perceptrones Multicap


```python
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

### Imágenes en formato RGB

#### Maquina de vectores de soporte

```python
estimators = [
    ('svc', SVC(random_state=SEED))
]

pipeline = Pipeline(estimators)

parameters = {
    'svc__kernel':['rbf', 'poly', 'sigmoid'],
    'svc__C': [0.1, 1, 10, 100],
    'svc__gamma': ['scale', 1, 0.1, 0.01, 0.001]
}

RGB_gs = GridSearchCV(pipeline,
                      parameters,
                      verbose=1,
                      cv = 3
                     )
RGB_gs.fit(RGB_train, RGB_y_train)
```

#### Perceptrones Multi capa

```python
estimators = [
    ('mlp', MLPClassifier(random_state=SEED,
                          max_iter=500
                         ))
]

pipeline = Pipeline(estimators)

parameters = {
    'mlp__hidden_layer_sizes': [
        (100,100,100,100,100,100),#5 capas
        (100,100,100,100,100,100,100,100,100,100,100,100), #10 capas
        (100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100)], #15 capas
    'mlp__activation': ['tanh', 'relu'],
    'mlp__solver': ['sgd', 'adam'],
    'mlp__alpha': [0.0001, 0.001],
    'mlp__learning_rate': ['constant','adaptive']
}

RGB_MLP_gs = GridSearchCV(pipeline,
                          parameters,
                          verbose=2,
                          cv = 3,
                          n_jobs=-1
                         )
RGB_MLP_gs.fit(RGB_train, RGB_y_train)
```

### Imágenes en escala de grises

#### Maquina de vectores de soporte

```python
estimators = [
    ('svc', SVC(random_state=SEED))
]

pipeline = Pipeline(estimators)

parameters = {
    'svc__kernel':['rbf', 'poly', 'sigmoid'],
    'svc__C': [0.1, 1, 10, 100, 1000],
    'svc__gamma': ['scale', 1, 0.1, 0.01, 0.001, 0.0001]
}

gray_gs = GridSearchCV(pipeline,
                      parameters,
                      verbose=1,
                      cv = 3
                     )
gray_gs.fit(gray_train, gray_y_train)
```

#### Perceptrones Multi Capa

```python
estimators = [
    ('mlp', MLPClassifier(random_state=SEED,
                          max_iter=500
                         ))
]

pipeline = Pipeline(estimators)

parameters = {
    'mlp__hidden_layer_sizes': [
        (100,100,100,100,100,100),#5 capas
        (100,100,100,100,100,100,100,100,100,100,100,100), #10 capas
        (100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100)], #15 capas
    'mlp__activation': ['tanh', 'relu'],
    'mlp__solver': ['sgd', 'adam'],
    'mlp__alpha': [0.0001, 0.001],
    'mlp__learning_rate': ['constant','adaptive']
}

gray_MLP_gs = GridSearchCV(pipeline,
                          parameters,
                          verbose=2,
                          cv = 3,
                          n_jobs=-1
                         )
gray_MLP_gs.fit(gray_train, gray_y_train)
```

## Evaluación

### Imágenes en formato RGB

#### Maquina de vectores de soporte

```python
RGB_y_pred = RGB_gs.predict(RGB_test)

RGB_score = classification_report(RGB_y_test, RGB_y_pred, digits=4)

print(RGB_score)

print('Mejores hiperparametros: ', RGB_gs.best_params_)

RGB_cnf_matrix = confusion_matrix(RGB_y_test, RGB_y_pred)
#cnf_matrix
list_confusion_matrix(RGB_cnf_matrix,class_name)
```

#### Perceptrones Multi Capa

```python
RGB_y_pred = RGB_MLP_gs.predict(RGB_test)

RGB_score = classification_report(RGB_y_test, RGB_y_pred, digits=4)

print(RGB_score)

print('Mejores hiperparametros: ', RGB_MLP_gs.best_params_)

RGB_cnf_matrix = confusion_matrix(RGB_y_test, RGB_y_pred)
#cnf_matrix
list_confusion_matrix(RGB_cnf_matrix,class_name)
```

### Imágenes en escala de grises

#### Maquina de vectores de soporte

```python
gray_y_pred = gray_gs.predict(gray_test)

gray_score = classification_report(gray_y_test, gray_y_pred, digits=4)

print(gray_score)

print('Mejores hiperparametros: ', gray_gs.best_params_)

gray_cnf_matrix = confusion_matrix(gray_y_test, gray_y_pred)
#cnf_matrix
list_confusion_matrix(gray_cnf_matrix,class_name)
```

#### Perceptrones Multi Capa

```python
gray_y_pred = gray_MLP_gs.predict(gray_test)

gray_score = classification_report(gray_y_test, gray_y_pred, digits=4)

print(gray_score)

print('Mejores hiperparametros: ', gray_MLP_gs.best_params_)

gray_cnf_matrix = confusion_matrix(gray_y_test, gray_y_pred)
#cnf_matrix
list_confusion_matrix(gray_cnf_matrix,class_name)
```

Al visualizar los resultados se puede identificar que ninguno de los modelos logró un gran desempeño, por tanto, se debe explorar un poco más las técnicas de clasificación de imágenes ya que solo se logró un poco más de 50% de la tasa de acierto. Como se puede evidenciar en las matrices de confusión, la más de la mitad de los datos quedaron bien clasificados. El modelo que obtuvo mejor desempeño fue las máquinas de vectores de soporte. Sin importar si la imagen estuviera en escala de grises o en formato RGB.
