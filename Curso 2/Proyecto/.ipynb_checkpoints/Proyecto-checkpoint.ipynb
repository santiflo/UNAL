# Proyecto de curso: Clasificador de medusas

## Entendimiento de los datos
El siguiente conjunto de datos fue extraido del repositorio www.kaggle.com, el cual cuenta con multiples imagenes sobre medusas 

```python
import cv2
import numpy as np
import os
from tqdm import tqdm
from matplotlib import pyplot as plt
```

```python
class_name = ['barrel_jellyfish', 'blue_jellyfish', 'compass_jellyfish', 'lions_mane_jellyfish', 'mauve_stinger_jellyfish', 'Moon_jellyfish']
class_names_label = {class_name: i for i, class_name in enumerate(class_name)}
count_class = len(class_name)
IMAGE_SIZE = (150,150)
```

```python
def load_data():
    
    # Directorio donde se encuentran almacenado los datasets
    dir = ['Train_Test_Valid/Train', 'Train_Test_Valid/test', 'Train_Test_Valid/valid']
    
    train = None
    test = None
    valid = None
    temp = None
    
    for dataset in dir:
        
        images = []
        labels = []
        print(f'Cargando "{dataset}"')
        
        for folder in os.listdir(dataset):
            
            label = class_names_label[folder]
            
            for file in tqdm(os.listdir(os.path.join(dataset, folder))):
                
                img_path = os.path.join(os.path.join(dataset, folder), file)
                image = cv2.imread(img_path)
                image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                image = cv2.resize(image, IMAGE_SIZE)
                temp = image

                images.append(image)
                labels.append(label)
        images = np.array(images, dtype = 'uint8')
        labels = np.array(labels, dtype = 'int32')

        if dataset[-1]=='n': train = [images, labels]
        elif dataset[-1]=='t': test = [images, labels]
        elif dataset[-1]=='d': valid = [images, labels]
    
    return train, test, valid, temp
```

```python
(train_X, train_y), (test_X, test_y), (valid_X, valid_y), temp= load_data()
```

```python
def display_random_image(class_names, X, y):
    index = np.random.randint(X.shape[0])
    plt.figure()
    plt.imshow(X[index])
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.title(f'Image #{index} : '+class_names[y[index]])
    plt.show()
```

```python
display_random_image(class_name, train_X, train_y)
```
